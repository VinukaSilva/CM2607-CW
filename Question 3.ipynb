# Part C

```python
# Array that stores the terms of the taylor series
ms = np.empty(50,dtype=object)     
xrange = np.linspace(5,-5,500) 
y = np.zeros([51,500])

ms[0] = f_x.subs(x,center_point)
f= sym.lambdify(x,ms[0], 'numpy')
y[0,:] = f(xrange)

for n in range(1,50):
  ms[n] = ms[n-1] + f_x.diff(x, n).subs(x, center_point) * (x - center_point)**n / sym.factorial(n)
  f = sym.lambdify(x, ms[n], 'numpy')
  y[n, :] = f(xrange)

f= sym.lambdify(x,f_x, 'numpy')
y[50,:] = f(xrange)
plt.plot(xrange, y[0,:])
plt.plot(xrange, y[4,:])
plt.plot(xrange, y[9,:])
plt.plot(xrange, y[50,:])
plt.ylim(-5, 5)

plt.legend(["1","5","10","50"])
plt.show()
```

# Part D

```python
# Define the symbolic variable and the series
x = sym.symbols('x')
f_x = sym.sin(x)
N = 16

# Taylor Series
ts = sym.series(f_x, x0=0, n=N).removeO()

# Substitute x = pi/4 into the Taylor series
approximation = ts.subs(x, sym.pi/4)

# Calculate the desired value using the approximation
result = sym.pi/2 * approximation

real_value = sym.pi/2 * sym.sin(sym.pi/4)

# Print the result
print("Real Value = ", real_value.evalf())
print("Approximation Result = ", result.evalf())

# Generate x values for plotting
x_vals = np.linspace(-2 * np.pi, 2 * np.pi, 500)

# Initialize an array to store the series values
y_vals_series = np.zeros_like(x_vals, dtype=float)

y_vals_sin = np.sin(x_vals)

# Evaluate the series up to the 4th order at each x value
for i in range(1, N):
    term = sym.series(f_x, x0=sym.pi/4, n=i).removeO()
    f_series = sym.lambdify(x, term, 'numpy')
    y_vals_series += f_series(x_vals)

# Plot the series and sin(x) on the same graph
plt.plot(x_vals, y_vals_series, label='Taylor Series (4 terms)')
plt.plot(x_vals, y_vals_sin, label='sin(x)')
plt.ylim(-8, 8)
plt.figure(figsize=(12, 6))     # Figure size of the plot
```
