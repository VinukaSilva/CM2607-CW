```python
import sympy as sym
import math
import numpy as np
import matplotlib.pyplot as plt
```

# Part A

```python
x = sym.symbols('x')

f_x = x * sym.sin(x/4)

f = sym.lambdify(x, f_x, 'numpy')

# Generating x values range
x0 = np.arange(-6*np.pi, 3*np.pi, 0.01)

y = f(x0)

# Create the plots
plt.figure(figsize=(12, 6))

# Plot the function
plt.plot(x0, y, label='f(x)')
plt.title('Function')
plt.legend()
```

# Part B

```python
# Defining the function 
f_x = sym.sin(x)

# Setting the center point of the Taylor Series
center_point = sym.pi/4

# Terms for the Taylor Series
n_th_terms = 50

# Taylor Series
# parameter1 = function
# parameter2 = function variable
# parameter3 = center value of the series
# parameter4 = the terms of the series 
ts = sym.series(f_x, x, center_point, n_th_terms)
ts
```

# Part C

```python
# Array that stores the terms of the taylor series
ms = np.empty(50,dtype=object)     
xrange = np.linspace(5,-5,500) 
y = np.zeros([51,500])

ms[0] = f_x.subs(x,center_point)
f= sym.lambdify(x,ms[0], 'numpy')
y[0,:] = f(xrange)

for n in range(1,50):
  ms[n] = ms[n-1] + f_x.diff(x, n).subs(x, center_point) * (x - center_point)**n / sym.factorial(n)
  f = sym.lambdify(x, ms[n], 'numpy')
  y[n, :] = f(xrange)

f= sym.lambdify(x,f_x, 'numpy')
y[50,:] = f(xrange)
plt.plot(xrange, y[0,:])
plt.plot(xrange, y[4,:])
plt.plot(xrange, y[9,:])
plt.plot(xrange, y[50,:])
plt.ylim(-5, 5)

plt.legend(["1","5","10","50"])
plt.show()
```

# Part D

```python
# Defining the x variable and the function
x_sym = sym.symbols('x')
f_x = sym.sin(x_sym)

# Defining the x value ranges
x_vals = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x_vals))

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']
plt.figure(figsize=(12, 8))

# Plotting the approximations of the taylor series 
for i, label in zip(range(4), labels):
    y = y + (f_x.diff(x_sym, i).subs(x_sym, sym.pi / 4)) / (sym.factorial(i)) * (x_vals - sym.pi / 4) ** i
    plt.plot(x_vals, y, label=label)

plt.plot(x_vals, np.sin(x_vals), 'k', label='Analytic')
plt.grid()
plt.title('Taylor Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

x = np.pi/4
y = 0

# Calculating the numeric value
for n in range(5):
    y = y + (np.pi/2) * ((-1)**n * (x)**(2*n+1)) / sym.factorial(2*n+1)
    
# Calculating the real value of the given approximation 
real_value = sym.pi/2 * sym.sin(sym.pi/4)

# Print the result
print("Real Value = ", real_value.evalf())
print("Approximation Result = ", y)
```
