```python
import sympy as sym
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
```

# Part A

```python
x = sym.symbols('x')

logistic = 1 / (1 + sym.exp(-x))

f = sym.lambdify(x, logistic, 'numpy')

# Generating x values range
x0 = np.linspace(-2*np.pi, 2*np.pi, 1000)

y = f(x0)

# Create the plots
plt.figure(figsize=(12, 6))

# Plot the function
plt.plot(x0, y, label='f(x)', color = 'g')
plt.title('Logistic Function')
plt.legend()
```

# Part B

```python
diff_f_x = sym.diff(logistic, x)
simplify(diff_f_x)
```

```python
f1 = sym.lambdify(x, diff_f_x, 'numpy')

# Generating x values range
x0 = np.linspace(-2*np.pi, 2*np.pi, 1000)

y = f1(x0)

# Create the plots
plt.figure(figsize=(12, 6))

# Plot the function
plt.plot(x0, y, label='f(x)', color = 'y')
plt.title('Derivative of Logistic Function')
plt.legend()
```
