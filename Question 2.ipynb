```python
import numpy as np
import scipy.fftpack as sfft
import matplotlib.pyplot as plt
```

For demonstrating aliasing in DFT, I have used a sin function whose frequency is close to half the sampling rate, causing aliasing to occur. The function which I have chosen is; 
                                $$sin(x)+0.8 \ sin(15x)$$

```python
x1 = np.arange(-8,8, 0.01) 
x2 = np.arange(-8,8, 0.1) 

y1 = np.sin(x1) + 0.8*np.sin(15*x1) 
y2 = np.sin(x2) + 0.8*np.sin(15*x2) 

# DFT
yf1 = sfft.fft(y1)
yf2 = sfft.fft(y2)

# IDFT
yf1_ = sfft.ifft(yf1)
yf2_ = sfft.ifft(yf2)

plt.figure(figsize=(10, 6))
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x1,np.real(yf1_))
plt.plot(x2,np.real(yf2_))
plt.legend(["100Hz","10Hz","IFFT 100Hz","IFFT 10Hz"])
plt.show()
```
