```python
import numpy as np
import scipy.fftpack as sfft
import matplotlib.pyplot as plt
```

For demonstrating aliasing in DFT, I have used a sin function whose frequency is close to half the sampling rate, causing aliasing to occur. The function which I have chosen is; 
                                $$sin(x)+0.45 \ sin(50x)$$

```python
x = np.arange(-np.pi,np.pi, 0.001)
x1 = np.arange(-np.pi,np.pi, 0.1)

y = np.sin(x) + 0.45*np.sin(50*x)
y1 = np.sin(x1) + 0.45*np.sin(50*x1)

# DFT
yf = np.fft.fft(y)
yf1 = np.fft.fft(y1)

# DFT Plot
plt.plot(x,yf)
plt.plot(x1,yf1)
plt.legend(["1000Hz","10Hz"])
plt.show()

# Function Plot
plt.plot(x,y)
plt.plot(x1,y1)
plt.legend(["1000Hz","10Hz"])
plt.show()
```
